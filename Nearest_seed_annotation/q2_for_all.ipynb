````python
def produce_csv(in_lst):
    for ft_qza in in_lst:
        a_ft_table = Artifact.load('qiime2/' + ft_qza + '/' + ft_qza + '-dada2_table.qza')
        
        a_biom = feature_table.methods.filter_features(table = a_ft_table,
                                               min_frequency = 1)
        a_biom_table = a_biom.filtered_table
        
        biom_table = a_biom_table.view(biom.Table)
        
        biom_table_pd = a_biom_table.view(pd.DataFrame)
        
        biom_table_pd.to_csv('qiime2/' + ft_qza + '/' + ft_qza + '-ft.csv')
        
def split_list(input_list, n):
    k, m = divmod(len(input_list), n)
    return [input_list[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n)]
````
